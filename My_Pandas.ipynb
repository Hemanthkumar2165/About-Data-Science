```python
#filling with missing in null place
df["age"].fillna("Missing",inplace = True)
#df
```

```python
#ffill -> null value is filled with above value     #limit is a parameter 
df["class"].fillna(method = "ffill",inplace = True)
df["marks"].fillna(method = "ffill",limit=1,inplace = True)
df
```

```python
#replace a particular value , we can change one or multiple
df.replace(to_replace=["A"] , value = "a")
df.replace(to_replace=["Missing"] , value = 17)
```

```python
dw = pd.DataFrame(np.random.randn(6,3),columns = ['first','second','third'])
print(dw)
print(dw.apply(np.mean))
```

```python
#covariance -> measures the direction of the relationship between two variables
a1 = pd.Series(np.random.randn(8))
a2 = pd.Series(np.random.randn(8))
print(a1.cov(a2))
```

```python
b = pd.DataFrame(np.random.randn(8,4),columns=['a1','a2','a3','a4'])
print(b['a1'].cov(b['a2']))
print(b.cov())
```

```python
#correlation -> shows lineaer relationship blw two array of values
print(b['a1'].corr(b['a2']))
print(b.corr())
#if any non-numeric column is excluded automatically
```

```python
#rolling func -> can be applies to series of data and #expanding -> is same i.e.,int the place of window we write min_periods
c = pd.DataFrame(np.random.randn(10,4),
                 index = pd.date_range('1/1/2000',periods=10),
                 columns = ['one','two','three','four'])
print(c.rolling(window=3).mean())
```

```python
#melt() -> is to convert wide dataframe to narrow ones
f = pd.DataFrame({'A':['a','b','c'],
                  'B':[4,5,7],
                  'C':[9,3,5]})
print(f)
pd.melt(f, id_vars=['A'] , value_vars=['B','C'])      #we can also add var_name="mine",value_name="your"
```
